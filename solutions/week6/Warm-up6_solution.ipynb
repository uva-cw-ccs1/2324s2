1. Use the API querying function from The Guardian provided in the lecture, and the documentation available here https://open-platform.theguardian.com/documentation/, to create a custom query from the Guardian of your choosing. To complete this exercise, your custom query must return at least 20 rows of data. Use the API credential (key) provided in the function in the lecture notebook. If you receive an error regarding a conflicting query occurring at the same time, try to run the snippet again. You should then receive a valid response. 

2. Read the response from the API into a dataframe.

3. Make sure that all titles of articles are pre-processed (unnecessary spaces removed, lowercased).

```python
# This is an adapted version of the function - I choose the content endpoint
def query_api2(query, from_date, page, api_key):
    """
    Function to query the API for a particular tag
    returns: a response from API
    """
    response = requests.get("https://content.guardianapis.com/search?q=" # note the change here
                            + query + "&from-date=" + from_date 
                            +"&page=" + str(page) + "&page-size=200&api-key=" + api_key)
    return response
```

```python
#I provide it with the necessary arguments 
response = query_api2('climate', '2023-01-01', '1', 'e723ffce-dfd5-427e-b9b8-779f5efedb02')
response # all went well
```

```python
#let's have a look how the data looks like
response.json()
```

```python
#using json_normalize I create a dataframe. Note how I specify the values that json_normalize should use
df = pd.json_normalize(response.json()['response']['results'])
df.head()
```

```python
df.columns
```

```python
#Now I will do two operations on the titles - first remove spaces and then lowercase it. I do it in two steps and create a new column `webTitle_pre`
df['webTitle_pre'] = df['webTitle'].str.strip()
```

```python
df['webTitle_pre'] = df['webTitle_pre'].str.lower()
```
